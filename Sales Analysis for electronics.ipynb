#### Question 2: What city had the highest number of sales?

```python
#### Add a city column
# lets use .apply() method

#all_data['Column'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])
#all_data.head()

# to split the address. the x value is the cell content
#for each of the cell content we want to split it by the comma and grab the first index

def get_city(address):
    return address.split(',')[1] 
#we can replace this function in --->

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x))
all_data.head()

#remove the name 'Column' by
#all_data = all_data.drop(columns='Column')

all_data.head(5)
```

```python
#further split the address by state
#lets use .apply()

def get_state(address):
    return address.split(',')[2] #second index
#all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) +''+ get_state(x))
#all_data.head()

#do another split to get the zip code

def get_state(address):
    return address.split(',')[2].split(' ')[1] #the space between TX and 75001

#all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) +' '+ get_state(x))
#all_data.head()

#you can use fstring 
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f'{get_city(x)} ({get_state(x)})')
all_data.head()
```

```python
#city with the highest number of sales

results = all_data.groupby('City').sum()
results
```

```python
#San Francisco

import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks (cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show
```

### Question 3: What time should we display advertisements to maximize likelihood of customers buying the product?

```python
all_data.head()
```

```python
#look at the order date; aggregate the order date into their distribution over a 24hr period
#convert the order date to a datetime object

all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
#groupby the hour

all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data['Count'] = 1
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
 

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()

#all_data.groupby(['Hour']).count()
```

11 am - 12 am and 6 pm -7pm are the peak times to display

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)] 
#use shift + tab to check the meaning of functions
#df.head(20)

#group duplicate products
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df.head(20)
```

```python
#group duplicate products
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
#count the pairs of items that occur frequently
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

    #print(count)

#count.most_common(10)
for key, value in count.most_common(10):
    print(key,value)
```

### Question 5: What product sold the most and why do you think it sold the most?

```python
#sum up the quantity ordered based on grouping by the product

product_group = all_data.groupby('Product')

product_group.sum(numeric_only=True)
```

```python
product_group =all_data.groupby('Product')
quantity_ordered = product_group.sum(numeric_only=True)['Quantity Ordered']

products = [product for product, df in product_group]


plt.bar(products,quantity_ordered)
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```
